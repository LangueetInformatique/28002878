```python
def lire_fichier(chemin):
  f = open(chemin, "r", encoding="utf-8")
  chaine = f.read()
  f.close()
  return chaine
```

```python
loc_corpus1=lire_fichier("locations_corpus1.txt")
loc_corpus2=lire_fichier("locations_corpus2.txt")
```

```python
loc1=loc_corpus1.split(",")
loc2=loc_corpus2.split(",")
```

```python
list_loc_difference = []
for loc in loc2:
    if loc not in loc1:
        list_loc_difference.append(loc)

print(list_loc_difference) #tous les "locations" qui appraissent dans l'echantillion2 et pas dans l'echantillion1
```
